```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시여
%matplotlib inline

x=list(range(0,100,5))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]
plt.plot(x,y)

# 차트 꾸미기
plt.plot(x,y, "b--", label="first") #차트 선색 및 크기설정
plt.plot(x,y2, "r--", label="second") #차트 선색 및 크기설정
plt.legend(loc=2)
```

# 2차원 Subplot

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시여
%matplotlib inline

x=list(range(0,100,5))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(122)
plt.plot(x,y, 'g', lw=1.5, label='2nd')
plt.plot(x,y, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```
