```python
import numpy as np
import pandas as pd

kcsData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
kcsData.head(1)
```

# 실습(ma구하기-일반)

```python
kcsData.columns = kcsData.columns.str.upper()

sortedData = kcsData.sort_values(["REGIONID","PRODUCT"])
sortedData.reset_index(inplace=True, drop=True)
sortedData.head()

sortedData["MA"] = sortedData["QTY"].rolling(window=3,center=False,min_periods=1).mean()
sortedData.head()
```

# 실습(ma구하기-고급)

```python
def sub_func(data):
    data["MA2"] = data["QTY"].rolling(window=3,center=True,min_periods=1).mean()
    return data
```

```python
kcsData.columns = kcsData.columns.str.upper()

sortedData = kcsData.sort_values(["REGIONID","PRODUCT"])
sortedData.reset_index(inplace=True, drop=True)
sortedData.head()

gbData = sortedData.groupby(["REGIONID","PRODUCT"])
len(gbData)
```

```python
eachData = gbData.get_group(list(gbData.groups)[0])
eachData.head()
```

```python
gbReault = sortedData.groupby(["REGIONID","PRODUCT"]).apply(sub_func)
gbReault.head(10)
```

## -함수생성(다른방법)

```python
#basicData = eachData.reset_index(drop=True)
#basicData.head()
```

```python
def ma_rolling(eachData, term):
    basicData = eachData.reset_index(drop=True)
    basicData["MA"] = basicData["QTY"].rolling(window=term,center=True,min_periods=1).mean()
    return basicData
#bascData.head(2)
```

```python
len(eachData)
```

```python
period = 17
grResult = sortedData.groupby(["REGIONID","PRODUCT"]).apply(ma_rolling, period)
grResult
```

# 계절성 산출

```python
grResult["RATIO"] = grResult["QTY"]/grResult["MA"]
grResult["WEEK"] = grResult["YEARWEEK"].astype(str).str[4:]
grResult.sort_values(["YEARWEEK"])
```

# 실습(std구하기-고급)

```python
def sub_func(data):
    data["STD"] = data["QTY"].rolling(window=3,center=True,min_periods=1).std()
    return data
```

```python
sortedData = kcsData.sort_values(["REGIONID","PRODUCT"])
sortedData.reset_index(inplace=True, drop=True)
sortedData.head()

gbReault = sortedData.groupby(["REGIONID","PRODUCT"]).apply(sub_func)
gbReault.head(10)
```
