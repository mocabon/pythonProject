# 데이터 정렬하기

```python
import numpy as np
import pandas as pd

customerData = pd.read_csv("./dataset/customerdata.csv")
customerData.head(1)
```

```python
customerData.sort_values(["AVERAGEPRICE"],ascending=[True]) # << 이상태로는 customerData 에 적용되지 않음
customerData.head(10)
```

```python
# inplace를 통해 데이터 정렬을 적용한다.
customerData.sort_values(["AVERAGEPRICE"],ascending=[True], inplace=True)
customerData.head(10)
```

# 데이터 요약분석하기

```python
# 평균구하기
meanValue = customerData.groupby(['CUSTTYPE','EMI'], as_index=True).mean()
meanValue.head()
#meanValue.reset_index(inplace=True)
```

# 실습(EMI별 MAX값)

```python
maxValue = customerData.groupby(['EMI'], as_index=True).max()
maxValue.head()
```

# 기본통계치 구하기

```python
stdValue = customerData.groupby(['CUSTID'], as_index=True).describe()
stdValue.head()
```
