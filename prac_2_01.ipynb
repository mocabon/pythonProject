# 실습(이상치 정제하기)

```python
import numpy as np
import pandas as pd
customerData = pd.read_csv("./dataset/customerdata.csv")
customerData.head(1)
```

```python
customerData["PRODUCT_AGE_NEW"]= np.where(customerData["PRODUCTAGE"] < 1, 1,\
                                 np.where(customerData["PRODUCTAGE"] < 2, 2,\
                                 np.where(customerData["PRODUCTAGE"] < 3, 3,\
                                 5,\
                                )))

customerData.head()
```

```python
customerData.head(2)
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5
```

```python
customerData["PRODUCT_AGE_NEW2"] = customerData.apply(refining, axis=1)
customerData.head()
```
