```python
import numpy as np
import pandas as pd

krmData = pd.read_csv("./dataset/kopo_region_mst.csv")
print(krmData.head(1))
kcsnData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv", encoding = "utf-8-sig") 
print(kcsnData.head(1))
```

```python
mergeResultR = pd.merge(krmData, kcsnData,\
        left_on="REGIONID", right_on = "REGIONID", how="left")\
[["REGIONID","REGIONNAME","PRODUCT","YEARWEEK","QTY"]]
mergeResultR.head()

mergeResultL = pd.merge(krmData, kcsnData,\
        left_on="REGIONID", right_on = "REGIONID", how="right")\
[["REGIONID","REGIONNAME","PRODUCT","YEARWEEK","QTY"]]
mergeResultL.head()

mergeResultI = pd.merge(krmData, kcsnData,\
        left_on="REGIONID", right_on = "REGIONID", how="inner")\
[["REGIONID","REGIONNAME","PRODUCT","YEARWEEK","QTY"]]
mergeResultI.head()
```

```python
# lest 와 right 의 차이에 따라 데이터 삭제가 발생
print("right join:: " + str(len(mergeResultR)))
print("left join:: " + str(len(mergeResultL)))
print("inner join:: " + str(len(mergeResultI)))
print(len(mergeResultL)-len(mergeResultI))
```
